```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.arange(0,3*np.pi,0.1)
y=np.cos(x)
plt.plot(x,y)
plt.savefig("filename.png",dpi=300,format="png")
```

```python
n=1024
x=np.random.normal(0,1,n) #(平均數,標準差,點數)
y=np.random.normal(0,1,n)
```

```python
t=np.arctan2(y,x)
plt.scatter(x,y, s=75, c=t, alpha=.5)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```
