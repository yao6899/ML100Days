```python
import pandas as pd
```

```python
dataframe=pd.DataFrame([['green','M','male','short'],
                       ['red','L','female','normal'],
                       ['blue','XL','male','long']])
```

```python
dataframe.columns=['color','size','sex','lenght']
```

```python
dataframe
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
dataframe['size_label']=LabelEncoder().fit_transform(dataframe['size'].values)
```

```python
dataframe
```

```python
pf=pd.get_dummies(dataframe['color'])
#get_dummies() 把資料表中的每個類別 對應的欄位
#經過 One-hot Encoding(一位有效編碼)
```

```python
dataframe=pd.concat([dataframe,pf],axis=1)
```

```python
tempdata=pd.DataFrame([['2020-11-01',24.8],
                       ['2020-11-02',22.8],
                       ['2020-11-03',None],
                       ['2020-11-03',25]],columns=['date','current_temp'])
```

```python
tempdata
```

```python
tempdata.fillna(0)
```

```python
tempdata.fillna(tempdata.current_temp.mean())
```

```python
tempdata.fillna(tempdata.current_temp.median())
```

```python
#前(後)補值最常使用在金融上，
#有時候因為颱風天導致沒有開盤，
#這時沒開盤那天的數值空了通常都會補前一天的價錢。

#函式一樣使用 fillna()，
#我們只需要進一步運用參數 method=‘ffill’
#即可填補前一列數值， method=‘bfill’ 填補後一列數值
```

```python
tempdata.fillna(method='bfill')
```
