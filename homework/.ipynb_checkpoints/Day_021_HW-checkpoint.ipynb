```python
# Seaborn 可以直接把 PANDAS 的 dataframe 當成資料匯入
# 本日範例，我們以Seaborn 內建的 IRIS 資料集做範例
```

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib as plt
```

```python
import numpy as np
```

```python
df=sns.load_dataset('iris')
```

```python
df
```

```python
sns.boxplot(data = df)
```

```python
sns.boxplot(data = df, orient = "h")
```

```python
# 當一個或兩個正在研究的變數是分類的時，我們使用像條帶線()、swarmplot()等的圖。

# 查看到每個物種petal_length的差異。但是，散點圖的主要問題是散點圖上的點重疊。
```

```python
sns.stripplot(x='species',y='petal_length',data=df) #stripplot 分類點散圖
```

```python
sns.stripplot(x='species',y='petal_length',data=df,jitter=True)
# 另一個可以用作「抖動」 的替代選項是函數群圖()。
# swarmplot此函數將散點圖的每個點都放在分類軸上，從而避免重疊點
```
