```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
```

```python
low=1
high=7
r=np.arange(low,high)
```

```python
probs=stats.randint.pmf(r,low,high) #.計算離散均勻分配的概率質量分佈 (probability mass function)
```

```python
plt.bar(r,probs)
plt.ylabel('P(X=x)')
plt.xlabel('x')
plt.title('pmf of DU(1,6)')
plt.show()
```

```python
cumsum_probs=stats.randint.cdf(r,low,high)## 3.計算離散均勻分配的累積機率 (cumulative density function)，pmf 的累加
plt.ylabel('P(X<=x)')
plt.xlabel('x')
plt.title('cdf of DU(1,6)')
plt.plot(r,cumsum_probs)
plt.show()
```

```python
# 5.產生符合隨機樣本點 (random sample)
X = stats.randint.rvs(low,high,size=20)
print(X)
plt.hist(X,bins=13)
plt.show()
#試試看，，每一次的結果一樣嗎?
```
